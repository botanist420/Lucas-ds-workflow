```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips, 
                hue="sex", size="size", sizes=(10, 100))
```

```python
sns.catplot(x="sex", y="total_bill", data=tips,
            col="day", col_wrap=2, kind="box")
```
